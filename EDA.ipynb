```python
df[['Open']].boxplot()
Q1 = np.percentile(df['Open'], 25,
                   method = 'midpoint')
Q3 = np.percentile(df['Open'], 75,
                   method = 'midpoint')
IQR = Q3 - Q1
print("Shape: ", df.shape)
 
# Upper bound
upper = np.where(df['Open'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['Open'] <= (Q1-1.5*IQR))
print("Total Number Of Outliers",len(upper[0])+len(lower[0]))
```

```python
#Extreme Upper bound
e_upper = np.where(df['Open'] >= (Q3+3*IQR))
#Extreme Lower bound
e_lower = np.where(df['Open'] <= (Q1-3*IQR))
print("Total Number Of Extreme Outliers",len(e_upper[0])+len(e_lower[0]))
```
